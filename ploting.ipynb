## construction groups

```python
df3=pd.read_excel("TPE_news_panel_construction.xlsx")
construction_list=df3[df3["is_construction"]==1]["department"].unique()
construction_list
```

```python
construction_mask=df["source"].isin(construction_list)
constuction_group=df[construction_mask]
```

```python
#dataframme reframe
try:
    constuction_group.drop(columns='Unnamed: 0', inplace=True)
except Exception:
    pass
constuction_group=constuction_group.reset_index().drop(columns="index")
constuction_group
```

```python
frequency_const=pd.DataFrame(constuction_group.groupby("date")["title"].agg("count"))
frequency_const=frequency_const.reset_index()
frequency_const=frequency_const.rename(columns={"title":"construction_count"})
frequency_const
```

## non-construction

```python
non_construction_list=df3[df3["is_construction"]==0]["department"].unique()
non_construction_mask=df["source"].isin(non_construction_list)
non_constuction_group=df[non_construction_mask]
#dataframme reframe
try:
    non_constuction_group.drop(columns='Unnamed: 0', inplace=True)
except Exception:
    pass
non_constuction_group=non_constuction_group.reset_index().drop(columns="index")
non_constuction_group
```

```python
frequency_non_const=pd.DataFrame(non_constuction_group.groupby("date")["title"].agg("count"))
frequency_non_const=frequency_non_const.reset_index()
frequency_non_const=frequency_non_const.rename(columns={"title":"non_construction_count"})
frequency_non_const
```

## comparison groups

```python
df2=pd.read_excel("TPE_news_comparison_group.xlsx")
comparison_group_list=df2["conparison_dpt"]
comparison_mask=df["source"].isin(comparison_group_list) 
comparison_group=df[comparison_mask]
```

```python
#dataframme reframe
try:
    comparison_group.drop(columns='Unnamed: 0', inplace=True)
except Exception:
    pass
comparison_group=comparison_group.reset_index().drop(columns="index")
comparison_group
```

```python
frequency_comp=pd.DataFrame(comparison_group.groupby("date")["title"].agg("count"))
frequency_comp=frequency_comp.reset_index()
frequency_comp=frequency_comp.rename(columns={"title":"conparison_count"})
frequency_comp
```

# all the data

## construction vs. comparison

```python
merged_df_temp=frequency_comp.merge(frequency_const,on="date",how="outer")
```

```python
merged_df=merged_df_temp.merge(frequency_non_const,on="date",how="outer")
```

```python
merged_df=merged_df.fillna(0)
merged_df.tail()
```

```python
# Extract the year as an integer
merged_df['year'] = merged_df['date'].str.split('-').str[0].astype(int)

# Add 1911 to the year
merged_df['year'] += 1911

# Replace the year in the 'date' column
merged_df['date'] = merged_df['year'].astype(str) + '-' + merged_df['date'].str.split('-').str[1]+"-"+merged_df['date'].str.split('-').str[2]
merged_df.drop(columns="year", inplace=True)
```

```python
# Ensure the 'date' column is of type datetime

merged_df['date'] = pd.to_datetime(merged_df['date'],format="%Y-%m-%d")
merged_df
# Set the 'date' column as the index
merged_df.set_index('date', inplace=True)
merged_df=merged_df.sort_index()
merged_df=merged_df.reset_index()
```

```python
#data in 2014
merged_df.loc[(merged_df["date"]<"2015-01-01")&(merged_df["date"]>"2014-01-01"),"date"]
```

```python
#missing alot of data in 2015
merged_df[(merged_df["date"]<"2016-01-01")&(merged_df["date"]>"2015-01-01")]["date"].unique()
```

```python
#data in 2016
merged_df.loc[(merged_df["date"]<"2017-01-01")&(merged_df["date"]>"2016-01-01"),"date"]
```

```python
#data in 2017
merged_df[(merged_df["date"]<"2018-01-01")&(merged_df["date"]>"2017-01-01")]
```

```python
daily_frequency=merged_df.reset_index()
```

```python
import plotly.express as px

fig = px.line(daily_frequency, x="date", y=["conparison_count","construction_count"], title='All data construction vs. comparison by days')
fig.add_vline(x="2018-11-24", line_width=3, line_dash="dash", line_color="green")
fig.add_vrect(x0="2014-12-25", x1="2018-11-23", 
              annotation_text="First term", annotation_position="top left",
              fillcolor="green", opacity=0.25, line_width=0)
fig.add_vrect(x0="2018-11-24", x1="2022-12-25", 
              annotation_text="second term", annotation_position="top left",
              fillcolor="yellow", opacity=0.25, line_width=0)
fig.show()
```

```python
# Resample to group by week and aggregate by count
monthly_frequency = merged_df.resample('M').sum()
monthly_frequency
```

```python
#  reframe the Dataframe
monthly_frequency=pd.DataFrame(monthly_frequency)
monthly_frequency=monthly_frequency.reset_index()

fig = px.line(monthly_frequency, x="date", y=["conparison_count","construction_count"], title='All data construction vs. comparison by months')
fig.add_vline(x="2018-11-24", line_width=3, line_dash="dash", line_color="green")
fig.add_vrect(x0="2014-12-25", x1="2018-11-23", 
              annotation_text="First term", annotation_position="top left",
              fillcolor="green", opacity=0.25, line_width=0)
fig.add_vrect(x0="2018-11-24", x1="2022-12-25", 
              annotation_text="second term", annotation_position="top left",
              fillcolor="yellow", opacity=0.25, line_width=0)
fig.show()
```
